```python
# The code is runned by Colab
import albumentations as A
import cv2
from google.colab.patches import cv2_imshow
import random

random.seed(7)

transform1 = A.HorizontalFlip(p=1)
transform2 = A.ShiftScaleRotate(p=1)
transform3 = A.Blur(blur_limit=50, p=1)
transform4 = A.InvertImg(p=1)
transform5 = A.GaussNoise(p=1, var_limit=[0, 80])
transform6 = A.RGBShift(p=1)

# Read an image with OpenCV and convert it to the RGB colorspace
image = cv2.imread("TB.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Augment an image
transformed_image1 = transform1(image=image)["image"]
transformed_image2 = transform2(image=image)["image"]
transformed_image3 = transform3(image=image)["image"]
transformed_image4 = transform4(image=image)["image"]
transformed_image5 = transform5(image=image)["image"]
transformed_image6 = transform6(image=image)["image"]
```

```python
cv2_imshow(transformed_image1)
cv2_imshow(transformed_image2)
cv2_imshow(transformed_image3)
cv2_imshow(transformed_image4)
cv2_imshow(transformed_image5)
cv2_imshow(transformed_image6)
```
